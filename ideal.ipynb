```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

def calculate_linear_regression(x_values, y_values):
    # Create a linear regression object
    regression = LinearRegression()
    
    # Reshape the x_values array to match the expected input shape
    x_values = np.array(x_values).reshape(-1, 1)
    
    # Fit the regression model using the provided data
    regression.fit(x_values, y_values)
    
    # Retrieve the slope (coefficient) and y-intercept of the line
    slope = regression.coef_[0]
    y_intercept = regression.intercept_
    
    # Return the equation for the line of best fit
    equation = f'y = {slope:.2f}x + {y_intercept:.2f}'
    return equation, slope, y_intercept

# Example usage
x_values = [n_01/volume,n_02/volume,n_03/volume,n_04/volume,n_05/volume]
y_values = press_array
equation, slope, y_intercept = calculate_linear_regression(x_values, y_values)

# Generate a range of x-values for plotting the line
x_plot = np.linspace(min(x_values), max(x_values), 100)
y_plot = slope * x_plot + y_intercept

# Plot the original data points
plt.scatter(x_values, y_values, color='red', label='Data Points')

# Plot the line of best fit
plt.plot(x_plot, y_plot, color='blue', label='Line of Best Fit')

# Add labels and a legend to the plot
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.title('Linear Regression')
plt.legend()

# Display the equation on the plot
plt.text(0.5, 8, equation, fontsize=12, bbox=dict(facecolor='yellow', alpha=0.5))

# Show the plot
plt.show()
```
